```python
from unittest.mock import inplace

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # visualization
import matplotlib.pyplot as plt
import re
%matplotlib inline


data = pd.DataFrame( pd.read_excel('./breast-cancer.xls') )  # Data Import from local/ Drive file



data.at[1 , 'age'] = 10
data.sample(len(data))



def range_averager(value):
    if re.search(r"^[0-9]{1,2}-[0-9]{1,2}$", str(value)) is not None:
        array = np.array(re.findall(r"[0-9]{1,2}", value), dtype=int)
        return np.median(array)
    elif  re.search( r"^[0-9]{1,2}$",  str(value))is not None:
        return value
    else:
        return False
print(data.loc[1, 'age'])
print(range_averager(data.loc[1, 'age']))



def uniques_translator(column):
    column_uniques = pd.unique(data[column])
    return column_uniques

print(uniques_translator('age'))
```
